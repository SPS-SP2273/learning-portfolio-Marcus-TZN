# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt

x = [i for i in range(0,7)]
y1 = [j for j in range(0,13,2)]
y2 = [k for k in range(0,19,3)]

plt.plot(x,                      # x-values
         y1,                     # y-values
         label = 'Y1 Line',      # line-label
         color = 'red',          # line-color
         linestyle = 'dashed'    # line-style
        )
plt.xlabel('x-values')           # x value labels
plt.ylabel('y-values')           # y value labels
plt.title('X vs Y')              # Title of the entire plot 
plt.grid(alpha=.25)              # Transparency of grid
plt.legend(loc='upper left')     # Location of the legend

plt.show()
```

### Things to note

Note: plt.legend only acceps the following arguments:
loc = supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

The order of which keyword arguments are specified does not matter

Abbreviations
color = c
linestyle = ls
linewidth = lw

So:

plt.plot(x,y, color='red', linestyle = 'dashed', linewidth = 2) 

is the same as

plt.plot(x,y, c = 'red', ls = 'dashed', lw = 2)

Otherwise if we do not want to define so much parameters for thegraph, we can just use default values

plt.plot(x,y) will do. 

## 1.2 Adding another plot

```python
from matplotlib import pyplot as plt

x = [i for i in range(0,7)]
y1 = [j for j in range(0,13,2)]
y2 = [k for k in range(0,19,3)]

plt.plot(x, y1, label = 'Y1 Line', 
         color = 'red', linestyle = 'dashed')

plt.plot(x, y2, label = 'Y2 Line',
        c = 'b')                        # b is blue

plt.xlabel('x-values')         
plt.ylabel('y-values')        
plt.title('X vs Y')
plt.grid(alpha=.25)              
plt.legend(loc='upper left')     

plt.show()
```

## 1.3 Yet another plot but with error bars

```python
from matplotlib import pyplot as plt

x = [i for i in range(0,7)]
y1 = [j for j in range(0,13,2)]
y2 = [k for k in range(0,19,3)]
y3 = [m for m in range(0,25,4)]
x_error, y_error = 0.1, 0.75

plt.plot(x, y1, label = 'Y1 Line', 
         color = 'red', linestyle = 'dashed')

plt.plot(x, y2, label = 'Y2 Line',
        c = 'b')                      

plt.errorbar(x, y3, xerr = x_error, yerr = y_error,     #xerr and #yerr are for error bars
         label = 'Y3 Line', c = 'green')

plt.xlabel('x-values')         
plt.ylabel('y-values')        
plt.title('X vs Y')
plt.grid(alpha=.25)              
plt.legend(loc='upper left')     

plt.show()
```

## 2.1 Adding mathematical functions

```python
x = [i for i in range(0,7)]
# x2 = x**2 will not work now, convert to array

np_x = np.array(x)
np_x2 = np_x**2
sin_x = np.sin(x)

plt.plot(np_x, np_x2,
        label = '$x^2$', c = 'r', ls = 'dashed')
plt.plot(np_x, sin_x, 
         label = '$sin(x)$', c = 'b')
plt.legend()

plt.show()
```

## We need another axis!

```python
np_x = np.linspace(0,10,10000)      #np.linspace(lower, upper, interval)

np_x2 = np_x**2
sin_x = np.sin(np_x)

plt.plot(np_x, np_x2,
        label = '$x^2$', c = 'r', ls = 'dashed')
plt.legend(loc = 'lower left')                # for y-axis 1

plt.twinx()                                   # new y-axis for the 
                                              # plot that comes after 

plt.plot(np_x, sin_x, 
         label = '$sin(x)$', c = 'b')
plt.legend(loc = 'lower right')

plt.show()
```

### Things to note

TLDR:
- With plt.twinx(), we now need two legends, one for each axis
- np.linspace will provide us with more points to create a smoother plot

# 4 A real example: Global Warming

## 4.1 Plotting data from files

We shall use the average temperature data for daily land from Berkeley Earth website

```python
import numpy as np
data = np.loadtxt('Complete_TAVG_daily.txt',
                 skiprows = 24)

date = data[:,0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

## 4.2 Styles

### fivethirtyeight

```python
import numpy as np
data = np.loadtxt('Complete_TAVG_daily.txt',
                 skiprows = 24)

date = data[:,0]
anomaly = data[:, -1]

plt.style.use('fivethirtyeight')          #The comic sans of matplotlib
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```
