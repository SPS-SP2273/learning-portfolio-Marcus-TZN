## Task 4

```python
# Now lets get to renaming the txt files in the folder
#Adjust the function abit

month_dictionary = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }


def rename_my_file(old_file_name):                        # Define this function again
    
    try:
        oldpath = os.path.join('.','os-collaboration-exercise-data',old_file_name)

        split1 = old_file_name.split('.')
        split2 = split1[0].split('-')

        extracted_month, date_and_location = split2

        mm = month_dictionary[extracted_month]
        new_name = f'{mm}-{date_and_location}.txt'

        oldpath = os.path.join('.','os-collaboration-exercise-data',old_file_name)
        newpath = os.path.join('.','os-collaboration-exercise-data',new_name)

        os.rename(oldpath, newpath)
        print(f'{old_file_name} is renamed to {new_name} successfully!')
        
    except:
        print(f'Unable to rename {old_file_name}! Check the file name if it has been renamed yet!')


Files = glob.glob('os-collaboration-exercise-data/**/*.txt', recursive=True)
for File in Files:
    Filename = os.path.split(File)[1]
    rename_my_file(Filename)
```

## Task 5

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
        'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

path = os.path.join('.','os-collaboration-exercise-data')

for city in cities:
    try:
        folder_path = os.path.join(path, city)
        os.mkdir(folder_path)
        print(f'Creating folder for {city}... Completed!')
    except:
        print(f'{folder_path} already exists...skipping file creation.')

        
```
